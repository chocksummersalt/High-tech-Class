```python
#2번 정수 n을 입력하고 n까지 짝수의 합
정수 = int(input("정수를 입력해주세요:"))
정수저장 = []
합 = 0
for i in range(1, 정수+1):
    if i% 2 == 0 :
     정수저장.append(i)
     합 = 정수 + i
print(합)
```
